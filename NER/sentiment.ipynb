```python
from cnsenti import Emotion
```

```python
import pandas as pd

train_dataframe = pd.read_csv('train_data_public.csv', index_col='id')
test_dataframe = pd.read_csv('test_public.csv')
train_dataframe.head()
```

```python
list(emotion.emotion_count(train_dataframe.loc[0, 'text']).values())[2:]
```

```python
emotion = Emotion()
train_data_feature = [list(emotion.emotion_count(train_dataframe.loc[i, 'text']).values())[2:]
                      for i in train_dataframe.index]
train_data_target = [train_dataframe.loc[i, 'class'] for i in train_dataframe.index]
train_data_target
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, Y_train, Y_test = train_test_split(train_data_feature, train_data_target, test_size=0.2,
                                                    random_state=0)
RFC = RandomForestClassifier()
RFC.fit(X_train, Y_train)
Y_predict = RFC.predict(X_test)
classification_report(Y_predict, Y_test)
```
